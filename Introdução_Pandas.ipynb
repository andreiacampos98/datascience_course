# JOIN

```python
# DataFrames de exemplo
df1 = pd.DataFrame({
    "ID": [1, 2, 3, 4],
    "Name": ["Alice", "Bob", "Charlie", "David"]
})

df2 = pd.DataFrame({
    "ID": [3, 4, 5, 6],
    "Score": [85, 90, 75, 88]
})

print(df1)

print(df2)
```

```python
# Left Join

left_merge = pd.merge(df1, df2, on="ID", how="left")

print(left_merge)
```

```python
left_merge.fillna(0)
```

```python
# Right Join

right_merge = pd.merge(df1, df2, on="ID", how="right")

print(right_merge)
```

```python
# Inner Join

inner_merge = pd.merge(df1, df2, on="ID", how="inner")

print(inner_merge)
```

```python
# Outer Join

outer_merge = pd.merge(df1, df2, on="ID", how="outer")

print(outer_merge)
```

# Join


```python
# DataFrames de exemplo
df1 = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie']}, index=[1, 2, 3])
df2 = pd.DataFrame({'Score': [85, 90, 95]}, index=[2, 3, 4])

print(df1)

```

```python
print(df2)
```

```python
 #Usando join para combinar com base no índice


joined_df = df1.join(df2, how='inner')  # Inner join
print(joined_df)

```
