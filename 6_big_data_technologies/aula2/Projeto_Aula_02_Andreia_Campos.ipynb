```
# File location and type
file_location = "/FileStore/tables/imdb_top_1000.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "false"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
 .option("inferSchema", infer_schema) \
 .option("header", "true") \
 .option("sep", delimiter) \
 .load(file_location)

display(df)
```

# SQL

```
df.createOrReplaceTempView("imdb_table")
```

## Task 01
Listar todos os filmes dirigidos por Christopher Nolan


```
%sql
SELECT Series_Title
FROM imdb_table
WHERE Director = 'Christopher Nolan';
```

## Taks 02
Contar o número de filmes por gênero

```
%sql
SELECT Genre, count("Genre") AS Nb_filmes
FROM imdb_table
GROUP BY Genre;
```

##Task 03 
Liste os 10 filmes com a maior pontuação no IMDB

```
%sql
SELECT Series_Title, IMDB_Rating
FROM imdb_table
ORDER BY IMDB_Rating DESC
LIMIT 10;
```

##Task 04
Conte o número de filmes por ano de lançamento

```
%sql
SELECT Released_Year, COUNT(Released_Year) AS Nb_filmes
FROM imdb_table
GROUP BY Released_Year;
```

##Task 05
Calcule a média de pontuação no IMDB por gênero

```
%sql
SELECT Genre, AVG(IMDB_Rating) AS Media_pontuacao
FROM imdb_table
GROUP BY Genre;
```

##Task 06
Liste os 10 filmes com o maior número de votos no IMDB

```
%sql
SELECT Series_Title, CAST(No_of_Votes AS INT) AS No_of_Votes_Int
FROM imdb_table
ORDER BY No_of_Votes_Int DESC
LIMIT 10;

```
