**Passo 3** - Criar uma nova variável **length_min_max** através da transformação **Min-Max** aplicada à variável **length**. Valida se os valores da nova variável são os **previstos pela transformação**.

```python
min_length = Carros3['length'].min()
max_length = Carros3['length'].max()

Carros3['length_min_max']= round(( Carros3['length']- min_length)  / (max_length - min_length),2)
```

**Passo 4** - Criar uma nova variável **width_std** através da **estandardização** da variável **width**. Validar se os novos valores da variável são os **previstos pela transformação**.

```python
avg = Carros3['width'].mean()
std = Carros3['width'].std()

Carros3['width_std'] = round(((Carros3['width'] - avg) / std)/3,2)
```

```python
Carros3.describe()
```

**Passo 5** - Criar uma nova variável **height_log** através de uma transformação **Logarítmica** aplicada a **height**. Validar se os novos valores da variável são os **previstos pela transformação**.

```python
import numpy as np
Carros3['height_log'] = round(np.log1p(Carros3['height']),2)
```
