# Toy Example

In this notebook, we are going to apply KMeans to a toy dataset.

**Step 1:** Load the dataset `toy_example.xlsx` using Pandas.

```python
import pandas as pd

df = pd.read_excel('toy_example.xlsx')
df.head(3)
```

**Step 2:** Apply KMeans from `sklearn.cluster` to our dataframe `df`. Define the number of clusters as 3. Check the labels of the clusters using the attribute `labels_`.

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=3).fit(df)
labels = model.labels_
labels
```

**Step 3:** Import `matplotlib.pyplot` as `plt`.

```python
import matplotlib.pyplot as plt
```

**Step 4:** Assign the column `Var1` to `xs`, and column `Var2` to `ys`.

```python
xs = df.iloc[:,0]
ys = df.iloc[:,1]
```

**Step 5:** Create a scatter plot of `xs` and `ys`, specifying the keyword argument `c=labels` to color the points by their cluster label.  You'll see that KMeans has done a good job of identifying the clusters!

```python
plt.scatter(xs, ys, c=labels)
```

**This is great**, but let's take it one step further and add the cluster centres (the "centroids") to the scatter plot.

**Step 6:** Obtain the coordinates of the centroids using the `.cluster_centers_` attribute of the `model`.  Assign them to `centroids`.

```python
centroids = model.cluster_centers_
centroids
```

**Step 7:** Assign column `0` of `centroids` to `centroids_x`, and column `1` of `centroids` to `centroids_y`.

```python
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
```

**Step 8:** In a single cell, create two scatter plots (this will show the two on top of one another).  Call `plt.show()` only once, at the end.

Firstly, repeat the scatter plot you made above.  Secondly, do a scatter plot of `centroids_x` and `centroids_y`, using `'X'` (a cross) as a marker by specifying the parameter `marker`. Set the size of the markers to be `200` using `s=200`.

```python
plt.scatter(xs, ys, c=labels)
plt.scatter(centroids_x, centroids_y, marker='X', s=200)
plt.show()
```

**Great work!** The centroids are important because they allow us to determine the average values of each variable for each cluster.
