##1. Cálculo de Taxa de Retenção
Descrição: Escreva um programa que calcule a taxa de retenção de funcionários.

Entrada: Total de funcionários no início do ano (inteiro) e total de funcionários no final do ano (inteiro).
Saída: Taxa de retenção (float, em porcentagem).

```python
inicio_ano = int(input("Total de funcionarios no inicio do ano: "))
final_ano = int(input("Total de funcionarios no inicio do ano: "))
taxa_retencao = final_ano/inicio_ano * 100

print(f"Taxa de retencao: {taxa_retencao:.2f}%" )
```

##2. Registro de Candidatos
Descrição: Use uma lista para registrar os nomes de candidatos que se inscreveram para uma vaga.

Entrada: Nomes dos candidatos inseridos pelo usuário.
Tarefa: Exiba a lista completa de candidatos inscritos.

```python
candidatos = []
while True:
  nome = input("Digita o nome do candidato (ou sair para finalizar): ")
  if nome.lower() =='sair':
    break
  candidatos.append(nome)
print("candidatos inscritos", candidatos)
```
