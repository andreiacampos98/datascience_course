##Exemplo 2: Clculo do lucro
Contexto: Voc vende produtos e deseja calcular o lucro baseado em um preo de venda e custo

```python
# Preo de venda e custo

preco_venda = 150.00
custo_produto = 100.00

# Calculo do lucro

lucro = preco_venda - custo_produto

#Exibir Resultado

print("Preo de venda: $", preco_venda )
print("Preo do produto: $", custo_produto )
print("Lucro: $", lucro )
```

##Exemplo 3: Desconto promocional
Contexto: Durante uma promoo, oferecemos um desconto de 20% no preo original.

```python
# Variaveis
preco_original = float(input("Insira o valor do preo: "))
desconto_percentual = 20

#calculo do preo com desconto

desconto = ( preco_original * desconto_percentual )/ 100
preco_final = preco_original - desconto

#Exibir valores

print("Preo original: $", preco_original)
print("Desconto: $", desconto)
print("Preo final", preco_final)


```



##Exemplo 4: Comparao de vendas
Contexto: Comparar as vendas de dois meses consecutivos para entender o desempenho.

```python
# Variaveis

mes1_vendas = int(input("Insira o valor das vendas no Mes 1: "))
mes2_vendas = int(input("Insira o valor das vendas no Mes 2: "))

#Comparacao
try:
  resultado_percentual = ((mes2_vendas - mes1_vendas)/(mes1_vendas))*100
except:
  print("An exception occurred")
#Exibicao
print("Resultado %: ", resultado_percentual, " %")

```

##Exemplo 5: Previso simples de crescimento
Contexto: Estimar as vendas no prximo ms baseado em um crescimento percentual esperado.

```python
# Vendas atuais e crescimento esperado

vendas_atuais = 1000

crescimento_percentual = 34

# Calculo das vendas esperadas

vendas_esperadas = vendas_atuais * (1+crescimento_percentual/100)

print("Vendas esperadas: ", vendas_esperadas)
```
