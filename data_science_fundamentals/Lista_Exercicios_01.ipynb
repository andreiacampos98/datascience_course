3.	Condicional Simples

Pergunte ao usuário se ele deseja comprar um móvel. Se ele responder "sim", mostre uma mensagem dizendo que o móvel foi reservado; caso contrário, informe que ele continua disponível.


```python
comprar=input("Deseja comprar um móvel? Se sim, escreva 'sim'. Caso contrário escreva 'nao'. ").lower()
if comprar=='sim':
  print("O seu movel foi reservado.")
elif comprar=='nao':
  print("Obrigada pela sua visita! Volte sempre!")
else:
  print("Não digitou sim nem não. Obrigada.")
```

4.	Condicional Composta

Peça ao usuário o preço de um móvel. Se o preço for superior a 1.000, exiba "Produto premium"; se for inferior ou igual a R$500, exiba "Produto econômico". Para os demais casos, exiba "Produto intermediário".

```python
preco=int(input("Indique o preço do móvel. "))
if preco > 1000:
  print("Produto premium")
elif preco > 500 and preco <= 1000:
  print("Produto intermediário")
elif preco <= 500:
  print("Produto económico")
```

5.	Loop com Contador

Exiba uma contagem regressiva de 5 segundos para anunciar um desconto em móveis.


```python
segundos=5
print("Contagem decrescente!!")
while segundos > 0:
  print(f"{segundos}")
  segundos-=1
  if segundos == 0:
    print("A época de descontos está oficialmente aberta")
```

6.	Loop com Entrada do Usuário

Crie um programa que permita ao usuário adicionar móveis a um carrinho de compras. O programa deve parar de pedir entradas quando o usuário digitar "sair".


```python
list=[]
while True:
  movel=input("Adicione os móveis que desejar ao carrinho de compras. Se nao quiser adicionar mais nenhum móvel, escreva 'sair'. ").lower()
  if movel=='sair':
    break
  list.append(movel)
print(f"No carrinho de compras tem os seguintes items: {list}")
```

7.	List Comprehension

Suponha que a loja tenha uma lista de preços de móveis. Crie uma nova lista com os preços com 10% de desconto usando list comprehension.


```python
moveis={"Sofá": 800, "Mesa":450, "Cadeira": 99, "Rack": 39, "Estante": 69}
novos_preco =[ f"{value * 0.9:.2f}" for value in moveis.values()]
print(f"Os novos preços são {novos_preco}")
```

8.	Slice

Considere uma lista de móveis em promoção: ["Sofá", "Mesa", "Cadeira", "Rack", "Estante"]. Mostre os 3 primeiros e os 2 últimos itens da lista.


```python
moveis_promocao= ["Sofá", "Mesa", "Cadeira", "Rack", "Estante"]
print(f"A lista de móveis em promoção são {moveis_promocao}")
primeiros_moveis=moveis_promocao[:3]
print(f"Os 3 primeiros móveis são {primeiros_moveis}")
ultimos_moveis=moveis_promocao[3:]
print(f"Os 2 últimos moveis são {ultimos_moveis}")
```

9.	Loop com Condicional

Dada uma lista de preços de móveis, exiba apenas os preços que são maiores precos = [1200, 850, 300, 500, 1500]


```python
precos=[1200, 850, 300, 500, 1500]
maiores_precos=[]
for preco in precos:
  if preco > 1000:
    maiores_precos.append(preco)

print(f"Os maiores preços sao: {maiores_precos}")
```

10.	Desafio: Soma de Valores

Crie um programa que pergunte ao usuário o preço de 5 móveis e calcule o valor total desses móveis. Use um loop para realizar a entrada de dados.


```python
i=0
total=0
while i < 5:
  preco = float(input("Por favor insera o preço de cada móvel: "))
  total +=preco
  i+=1
print(f"O preço total dos móveis é {total:.2f}€")
```
