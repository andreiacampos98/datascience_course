```python
import matplotlib.pyplot as plt

import pandas as pd
```

# 1. Linhas Simples

```python
# Dados

data = pd.Series([1, 3, 2, 5, 4])

print(data)
```

```python
#Gráfico

plt.plot(data, label = "Série 1", color="Green")
plt.legend()
plt.show()
```

#2. Barra Simples

```python
data = pd.Series([10, 20, 15, 25], index=["A", "B", "C", "D"])
data.plot(kind ="bar", label = "Valores", color ="skyblue")
plt.legend()
plt.show()
```

#3. Histograma com labels

```python
import numpy as np

data = pd.Series(np.random.randn(100))

data.plot(kind ="hist", bins=10, edgecolor="black", label="Distribuição", alpha=0.7)
plt.legend()
plt.title("Histograma com Labels")
plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.show()
```

#4. Gráfica de pizza com labels

```python
data = pd.Series([15, 25, 35, 25], index=["A", "B", "C", "D"])

print(data)
```

```python
data.plot(kind="pie", autopct ="%.2f%%", ylabel="", labels = ["Grupo A", "Grupo B", "Grupo C", "Grupo D"])
plt.title("Gráfico de Pizza com Labels")
plt.show()
```

# 5. Dois gráficos no mesmo eixo com labels

```python
x = range(1,6)
y1= [1,3,2,5,4]
y2 = [2,4,1,3,5]

plt.plot(x,y1, label ="Linha 1", marker="o", color="blue")
plt.plot(x,y2, label ="Linha 2", marker="x", color="red")
plt.title("Dois gráficos com labels")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.show()
```

```python
x = range(1,6)
y= [1,3,2,5,4]

plt.plot(x,y, marker="o", color="blue")
for i, value in enumerate(y):
    plt.text(x[i], y[i] + 0.2, f"{value}", ha="left")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.show()
```

# 8.Subplots com labels

```python
data1 = pd.Series([1, 3, 2, 5, 4])
data2 = pd.Series([10, 15, 8, 20])

print(data1)

print(data2)
```

```python
plt.figure(figsize = (10,5))

plt.subplot(2,1,1)
data1.plot(kind="line", label="Série 1", color ="orange")
plt.title("Subplot 1")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()

plt.subplot(2,1,2)
data2.plot(kind="bar", label="Série 2", color ="purple")
plt.title("Subplot 2")
plt.xlabel("Índices")
plt.ylabel("Valores")
plt.legend()

plt.tight_layout()
plt.show()
```
