##3. Pedidos at Finalizar (While)
Permita que o cliente faa vrios pedidos usando um while loop. O cliente digita "sair" para encerrar. No final, mostre todos os itens pedidos.

```python
pedido_cliente = []
while True:
  pedido=input("Escolha um prato:\nQuando terminar o seu pedido escreva 'sair'! ").lower()
  if pedido=="sair":
    break
  elif pedido in menu:
      print("pedido aceite")
      pedido_cliente.append(pedido)
  else:
    print(f"hoje no temos esse belo prato!\nPor favor escolha um dos seguintes pratos: {menu} ")

print(pedido_cliente)
for i in pedido_cliente:
  print(i)

print(f"Voc pediu : {' , '.join(pedido_cliente)}")
```

##5. Clculo do Total da Conta (For e If)
Cada item do menu tem um preo. Escreva um programa que calcule o total da conta baseado nos pedidos do cliente.

```python
menu= {
    "Baja Taco": 4.25,
    "Burrito": 7.50,
    "Bowl": 8.50,
    "Nachos": 11.00,
    "Quesadilla": 8.50,
    "Super Burrito": 8.50,
    "Super Quesadilla": 9.50,
    "Taco": 3.00,
    "Tortilla Salad": 8.00
}

def main ():
    print(menu)

    total=0
    try:
        while True:
            item=input("O que deseja? Se no pretender adicionar mais nenhum item por favor digite 'sair'. ").title()
            if item=='Sair':
              break
            if item in menu:
                total+=menu[item]
                print(f'Total a pagar: ${total:.2f}')
            else:
              print("O item que inseriu nao est disponivel. Por favor escolha um novo item.")
        print(f'Total a pagar: ${total:.2f}')
    except EOFError:
            print(f'total: ${total:.2f}')

main()

```

##10. Controle de Estoque (While, If e Dicionrio)
Crie um sistema que permita pedidos de clientes at que os ingredientes acabem. Atualize o estoque a cada pedido.

```python
stock= {
    "Baja Taco": 5,
    "Burrito": 2,
    "Bowl": 1,
    "Nachos": 2,
    "Quesadilla": 3,
    "Super Burrito": 2,
    "Super Quesadilla": 1,
    "Taco": 4,
    "Tortilla Salad": 5
}

def main ():
    print(menu)
    pedido={}
    total=0
    try:
        while True:
            item=input("O que deseja? Se no pretender adicionar mais nenhum item por favor digite 'sair'. ").title()
            if item=='Sair':
              break
            if item in menu:
              if stock[item] <= 0:
                print(f'No temos stock deste item {item}: {stock[item]}')
              else:
                total+=menu[item]
                if item not in pedido.keys():
                  pedido[item] = 1
                elif item in pedido.keys():
                  pedido[item] += 1
                print(f'Total a pagar: ${total:.2f}')
                stock[item] -= 1
            else:
              print("O item que inseriu nao est disponivel. Por favor escolha um novo item.")
        print(f'O seu pedido contm os seguintes items: {pedido}')
        print(f'Total a pagar: ${total:.2f}')
        print(stock)
    except EOFError:
            print(f'total: ${total:.2f}')

main()
```
