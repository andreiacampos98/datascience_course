```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fires/Fires.csv.zip', encoding='latin1')
```

```python
df.info()
```

---

```python
tmp = df.groupby('Region')['Total_hectares'].sum().sort_values().reset_index()
tmp
```

```python
plt.bar(tmp['Region'], tmp['Total_hectares'], color='orange')
```

```python
sns.barplot(x=df['Region'], y=df['Total_hectares'], color='orange', estimator='sum', errorbar=None)
```
