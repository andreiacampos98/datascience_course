```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
months = pd.date_range('2011-01-01','2011-11-01', freq='MS').strftime("%b").tolist()
poll = [9, 8.9, 8.8, 9.0, 9.1, 9.2, 9.1, 9.1, 9.1, 9.0, 8.6]
months, poll
```

```python
plt.figure(figsize=(10, 4.5))
plt.grid(color='b', linestyle='-', linewidth=.5, axis='y', alpha=.1)
plt.plot(months, poll)
plt.suptitle('Unemployment rate', x=.125, y=1, ha='left', size=15, weight='bold')
plt.title('Under president Obama', loc='left', fontdict={'color': 'grey'})
plt.xlabel('2011', fontdict={'fontweight': 'bold'})
plt.ylabel('%')
# plt.ylim(0, 12)
plt.ylim(8, 10)
plt.show()
```
