**6.	Join:**

*   Faça um join do DataFrame de vendas com o de gerentes, considerando o índice.


```python
# Quando fazemos o join considerando o indice, o que vai acontecer é a primeira linha do dataframe de vendas
# vai unir-se à primeira linha do dataframe de gerentes.
# Neste caso este join nao tem sentido lógico/de negócio.
vendas_gerentes_df = vendas_df.join(gerentes_df, how='left', lsuffix='_vendas', rsuffix='_gerentes')
display(vendas_gerentes_df)
```

```python
#Para fazer o join corretamente temos de definir o index

vendas_gerentes_join = vendas_df.set_index('ID Loja').join(gerentes_df.set_index('ID Loja'), how='left')
print(vendas_gerentes_join)
```

*   Explique o comportamento do join comparado ao merge.

***O join e o merge são usados para combinar dataframes. Por padrão, o join() utiliza os índices dos DataFrames para realizar a junção. Enquanto que o merge usa uma coluna do dataframe.***

**7. Análise de agrupamento:**


*   Agrupe os dados combinados por gerente e calcule:
      *   A soma total das vendas de cada gerente.
      *   A média de vendas por gerente.





```python
# Valor vendido por cada gerente
subset=all.groupby(all['Gerente'])['Valor Final'].sum().sort_values(ascending=False)
display(subset)
```

```python
# Média de valor vendido por cada gerente
all.groupby(all['Gerente'])['Valor Final'].mean().sort_values(ascending=False)
```

**8.	Identificação de padrões:**

*   Crie uma coluna derivada que calcule a percentagem de vendas de cada gerente em relação ao total de vendas.


```python
valor_gerente=pd.DataFrame(subset)
display(valor_gerente)
```

```python
valor_gerente['Percentagem']=round(((valor_gerente['Valor Final'] / valor_gerente['Valor Final'].sum()) *100), 2)
display(valor_gerente)
```
